```python
#import the random library (contains the randint built-in function)
import random as rand


#greet user to the game
print("Welcome to the Rock, Paper, Scissors Game!!!")

#create variable to store selection generated for the computer (rock, paper or scissors)
computer_selection = " "

#create variable that will store value for continuation of game by user (or not)
answer = 'y'



#create loop that will allow user to play game until the user decides not to play any longer
while answer != 'n' :

#ask user to enter a selection (rock, paper, or scissors)
#store selection into a string variable
    print(" ")
    user_selection = input("Enter rock, paper, or scissors:    ")
    
#convert user entry into all lower case letters for comparison against the computer-generated selection    
    user_selection_lower = user_selection.lower()


#print/display user selection    
    print(" ")
    print("User selects", user_selection)

#create random integer (1, 2, or 3) that will decide the object chosen by the computer
    computer_selection_random_number = rand.randint(1, 3)

#convert random integer selection to a string value or rock, paper or scissors    
    if computer_selection_random_number == 1 :
        computer_selection = "rock"
    elif computer_selection_random_number == 2 :
        computer_selection = "paper"
    else :
        computer_selection = "scissors"

#print/display computer selection        
    print("Computer selects", computer_selection)    
    
#compare selection(s) if user enters rock and the outcome
    if user_selection_lower == "rock" :
        if computer_selection == "rock" :
            print("rock & rock are the same. It's a draw.")
        elif computer_selection == "paper" :
            print("paper covers rock. You lose.")
        else :
            print("rock smashes scissors. You win!!!!")
        
#compare selection(s) if user enters paper and print the outcome        
    if user_selection_lower == "paper" :
        if computer_selection == "rock" :
            print("paper covers rock. You win!!!")
        elif computer_selection == "paper" :
            print("paper & paper are the same. It's a draw.")
        else :
            print("scissors cuts paper. You lose.")

#compare selection(s) if user enters scissors and print the outcome            
    if user_selection_lower == "scissors" :
        if computer_selection == "rock" :
            print("rock smashes scissors. You lose.")
        elif computer_selection == "paper" :
            print("scissors cuts paper. You win!!!")
        else :
            print("scissors & scissors are the same. It's a draw.")

#ask user if the user wants to continue playing game    
    print(" ")
    answer = input("Do you want to continue? y/n:    ")        

#if user leaves game, thank the user for playing    
print(" ")
print(" ")
print("Thank you for playing Rock, Paper, Scissors!!!")
```
